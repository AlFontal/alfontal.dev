## Introduction

During a lunch conversation with friends we ended up revisiting a childhood staple: "Guess Who?" (or "¿Quién es quién?" in Spanish). That quickly turned into an argument about the rules and, eventually, into a better question: is there a winning strategy, or is the game mostly luck?

![Cover of the Spanish version of the game from 1982. Source: todocoleccion.net](quien_es_quien_tapa.jpg)

This game is a classic board game where players ask yes/no questions to guess the identity of a character chosen by their opponent. The game is played with a set of characters, each with distinct features, and players take turns asking questions to narrow down the possibilities.

Given that setup, building a computational version of the game felt like a manageable side project, so I decided to try it.

The goal is to run many simulations with different parameters and strategies, then use the results to reason about the best way to play. We can save the strategic discussion for later and start with the implementation.

## Game Implementation

```python
@dataclass
class Character:
    name: str
    hair_color: str
    eye_color: str
    wears_glasses: bool
    has_beard: bool
    has_hat: bool
    has_scarf: bool

    def __repr__(self):
        return f"Character(name={self.name}, hair={self.hair_color}, eye_color={self.eye_color}," + \
               f" wears_glasses={self.wears_glasses}, has_beard={self.has_beard}," + \
               f" has_hat={self.has_hat}, has_scarf={self.has_scarf})"
    def __str__(self):
        return f"{self.name} ({self.hair_color}, {self.eye_color}, {self.wears_glasses}, " + \
               f"{self.has_beard}, {self.has_hat}, {self.has_scarf})"
```

```python
Character(
    name="Alice",
    hair_color="brown",
    eye_color="blue",
    wears_glasses=True,
    has_beard=False,
    has_hat=False,
    has_scarf=False
)
```
